## working with dates

```python
## converting to datetime dataype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1.extract year

```python
date['date_year']=date['date'].dt.year ## dt== date time

date.head()
```

## 2.extract month

```python
date['date_month']=date['date'].dt.month

date.head()
```

```python
date['date_month']=date['date'].dt.month_name()

date.head()
```

## 3.extract days

```python
date['date_days']=date['date'].dt.day

date.head()
```

```python
## day of week

date['date_days']=date['date'].dt.dayofweek

date.head()
```

```python
## day of week_name 

date['date_dow_name']=date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
## is weekend

date['date_is_weekend']= np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
## extract week of the year

date['date_week']=date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
## extract quarter
date['quarter']=date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

## extract time elapsed between dates

```python
import datetime

today=datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today-date['date']).dt.days
```

```python
## month passed

np.round((today-date['date'])/np.timedelta64(1,'M'),0)
```
