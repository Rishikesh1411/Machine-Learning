# 13 convertors

```python
pd.read_csv('matches.csv')
```

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
    
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('matches.csv',converters={'team2':rename})
```
