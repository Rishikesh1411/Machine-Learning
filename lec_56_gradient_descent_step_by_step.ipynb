```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
## Lets apply ols(ordinary least square)
from sklearn.linear_model import LinearRegression
```

```python
reg  = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red')
```

```python
### lets apply Gradient Descent assuming slope is m = 78.35
### and let's assume the starting value for intercept is b=0
y_pred  = ((78.35* X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
#X.ravel()-- covert an array into one dim array
loss_slope = -2 * np.sum(y-m*X.ravel()-b)  ## slope calculation for calculate new value of b
loss_slope
```

```python
## let's take learning rate = 0.01
lr = 0.1

step_size= loss_slope*lr ## (b_new=b_old - (learning_rate)*(slope at b initial))
step_size
```

```python
##calculating the new intercept
b = b - step_size
b
```

```python
y_pred1  = ((78.35 * X) + b ).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='green',label='b= {}'.format(b))
plt.plot(X,y_pred,color='blue',label='b=0',)
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Iteration 2
loss_slope = -2 * np.sum(y - m * X.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((m * X)+ b ).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='black',label=f'b = {b} ')
plt.plot(X,y_pred1,color='green',label='b= -20.927763408209216')
plt.plot(X,y_pred,color='blue',label='b=0',)
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Iteration 3
loss_slope  = -2  * np.sum(y - m* X.ravel()-b)
loss_slope
```

```python
step_size= loss_slope * lr
step_size
```

```python
b = b- step_size
b
```

```python
y_pred3 = ((m * X)+ b).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='orange',label=f'b = {b} ')
plt.plot(X,y_pred2,color='black',label='b= 25.11331608985106')
plt.plot(X,y_pred1,color='green',label='b= -20.927763408209216')
plt.plot(X,y_pred,color='blue',label='b=0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#### with new inital b = -100

b = -100
m = 78.35
lr=0.1

epochs =100

for i in range(epochs):
    loss_slope = -2 * np.sum( y - m * X.ravel() - b)
    stepsize = loss_slope * lr
    b = b - stepsize
    
    y_pred = m *X + b
    
    plt.plot(X,y_pred)
    
plt.scatter(X,y)
```
